```python
import numpy as np

def predict_next_word(model, tokenizer, text, seq_length=14):
    """
    Given some text, predicts the next word using the trained LSTM model.
    """
    # 1 Convert text to sequence of integers
    sequence = tokenizer.texts_to_sequences([text])[0]

    # 2 Pad or trim the sequence to match training length
    if len(sequence) < seq_length:
        sequence = [0]*(seq_length - len(sequence)) + sequence  # pad at start
    else:
        sequence = sequence[-seq_length:]  # keep last seq_length tokens

    sequence = np.array(sequence).reshape(1, seq_length)

    # 3 Predict probabilities for next word
    pred_probs = model.predict(sequence, verbose=0)

    # 4 Get index of most likely next word
    predicted_index = np.argmax(pred_probs, axis=1)[0]

    # 5 Convert index back to word
    for word, index in tokenizer.word_index.items():
        if index == predicted_index:
            return word

# Example usage:
text = "I want to meet today but i am busy until"
next_word = predict_next_word(model, tokenizer, text)
print(f"Next word suggestion: {next_word}")
# I want to meet today but I am busy until evening.
```
